```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the range for X and Y values
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)

# Create a meshgrid from the X and Y ranges
X, Y = np.meshgrid(x_range, y_range)

# Calculate the Z values based on the function Z = X**2 + Y**2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Z = X**2 + Y**2')

# Set the viewing angle for a better visualization
ax.view_init(elev=30, azim=30)

# Show the plot
plt.show()
```

```python
import pandas as pd
```

```python
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

show_data = data.loc[:, selected_columns]

print(show_data)
```

#### # Display the first 20 samples of the new DataFrame

```python
print(show_data.head(20))
```

```python
mean_values = show_data.mean()

std_values = show_data.std()

quartile_values = show_data.quantile([0.25, 0.5, 0.75])

print("Mean values:")
print(mean_values)

print("\nStandard deviation values:")
print(std_values)

print("\nQuartile values:")
print(quartile_values)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 4, figsize=(16, 4))

for i, feature in enumerate(['N_total', 'N_bulk', 'N_surface', 'R_avg']):
    axes[i].hist(show_data[feature], bins=20, edgecolor='black', color='skyblue')
    axes[i].set_title(f'{feature} Histogram')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
